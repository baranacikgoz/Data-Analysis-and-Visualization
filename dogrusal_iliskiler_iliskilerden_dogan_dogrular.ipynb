```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x = "total_bill", y = "tip", data = df);
```

```python
#dogrunun yanindaki acik mavi renkler sapmalari hata paylarini ifade ediyor
#ornegin 10-20 arasi degerler cok yakin oldugu icin acik mavilikler neredeyse hic yok
#ancak 40-50 arasi degerlerin uzakligindan dolayi acik mavilikler cok
```

```python
sns.lmplot(x = "total_bill", y = "tip", hue= "smoker",data = df);
```

```python
#cok guzel dogrular cikti mesela sigara icmeyenlerde, odenen hesap arrtikca
#bahsis de guzel bir egimle artmis ama sigara icenlerde bahsis artis hizi daha az
#ancak durum boyle bile olsa en yuksek bahsisin sigara icenden gelmesi manidar
```

```python
sns.lmplot(x = "total_bill", y = "tip", hue= "smoker", col = "time", data = df);
```

```python
#grafigi sabah ve aksam olarak iki sutun olarak iki defa olusturduk,
#sabah sigara icen/icmeyenlerin odedigi hesap miktarina gore verdigi bahsisin artisi
#ile aksamki yakin mi, degilse de goze almaya degecek fark var mi gibi sorular
#sormamiz gerekir kendimize
```

```python
sns.lmplot(x = "total_bill", y = "tip", hue= "smoker", col = "time",row="sex" , data = df);
```
