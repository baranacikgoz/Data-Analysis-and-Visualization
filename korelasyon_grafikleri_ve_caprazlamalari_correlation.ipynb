```python
import seaborn as sns
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

## Scatterplot(Saçılım Grafikleri)

```python
sns.scatterplot(x = "total_bill", y="tip", data=df);
```

```python
#grafige gore odenen hesap arttikca tip de artmis mesela 50dolar hesapta 10dolar tip
#ayriyeten 10-20 dolar araligindaki hesap odemelerde artma olsa da 2-4 dolar tip arasında yigilmalar olmus
```

```python
sns.scatterplot(x = "total_bill", y="tip", hue="time", data=df);
```

```python
sns.scatterplot(x = "total_bill", y="tip",style = "time", hue="time", data=df);
#time degiskenindeki farkliliklarin seklini degistirdi birini yildiz yapti biri yuvarlak
```

```python
sns.scatterplot(x = "total_bill", y="tip",hue="day",style = "time",  data=df);
#bu sekilde, hem gunleri ekledik, hem de style=time sayesinde 
#gunlerin kendi icinde de sabah ya da aksam olmasına bagli sekil degisikligi oldu
```

```python
#mesela mavi daire = pazartesi sabah yemegi
#mesela kirmizi yildiz = cuma aksam yemegi
```

```python
sns.scatterplot(x = "total_bill", y="tip",size = "size", hue="day",style = "time",  data=df);
```

```python
#birinci size; scatterplot fonksiyonunun bir argumanı, onun eşiti olan diger siz
#ise verisetimizdeki masada insan miktarı olan size, 
#bu sayede insan sayisi arttikca daireler ya da yildizlar da buyudu
#bir grafikte bin tane degisken harmanladık okuması cok anlamlı, bilgi iceren bir
#grafige donustu
```
