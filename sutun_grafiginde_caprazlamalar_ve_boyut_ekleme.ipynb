Caprazlamalar

```python
import seaborn as sns
from pandas.api.types import CategoricalDtype
diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
cut_kategoriler = ["Fair","Good","Very Good","Premium","Ideal"]
df.cut = df.cut.astype(CategoricalDtype(categories = cut_kategoriler, ordered = True))
df.head()
```

```python
sns.catplot(x = "cut", y = "price", data = df);
#catplot kategorik degiskenlerin sutun grafigi icin kullanılır
#burda cut degiskeni ile priceı caprazladik
```

```python
sns.barplot(x = "cut", y = "price", hue = "color", data = df);
#hue = boyut olarak ekler, boyut olarak color degiskeninini ekledik
```

```python
#normalde price 18000e kadar vs cikiyordu ancak seaborn hi-level bir kutuphane
#oldugu icin buyuk resmi gostermek adina iki kategorik degisken
#bir arada bulununca olan price degerinin ortalamasini gosterdi
#ustteki cubuklar da standart sapma degerleri
```

```python
#elde ettigimiz sonuclari dogrulamak;

df.groupby(["cut", "color"])["price"].mean()
```
