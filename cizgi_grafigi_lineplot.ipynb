```python
import seaborn as sns
fmri = sns.load_dataset("fmri")
df = fmri.copy()
df.head()
```

```python
df.groupby("timepoint")["signal"].describe()
```

```python
sns.lineplot(x = "timepoint", y = "signal", data=df);
```

* Tek olan koyu mavi çizgi, ortalama değere karşılık geliyor.
Yani timepoint 11 civarlarında iken, o zaman aralığına ait 56 tane sinyalin ortalaması -0.05miş.

* Etrafındaki açık mavilikler ise, standart sapmalardan oluşan güven aralıklarını belirtiyor.

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "event" ,data=df);
```

Burdan şu sonuç çıkıyor, event eklenmemiş olan ilk çizgi grafiğimizdeki peak yapma noktasını "stim" eventi etkilemiş

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "event" ,style= "event" ,data=df);
```

```python
sns.lineplot(x = "timepoint", 
             y = "signal", 
             hue = "event" ,
             style= "event" ,
             markers = True, dashes = False,
             data=df);
```

```python
sns.lineplot(x = "timepoint", 
             y = "signal", 
             hue = "region" ,
             style= "event" ,
             markers = True, dashes = False,
             data=df);
```

Normalde bir boyut daha eklemek isterken hue kullanıyoruz ve hue 1 boyut daha ekleyebiliyor. Ancak hue'da eklenen ayrı bir boyut var iken, style argumanında da hue'dekinden farklı bir değişken yazarsak örn: (hue=region, style=event) daire ya da yıldız şeklinde gelen bir boyut daha eklemiş gibi çok daha ayrıntılı gözlem yapabiliyoruz.
