```python
import seaborn as sns
diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
df.head()
```

```python
sns.kdeplot(df.price, shade = True);
```

```python
(sns.
 FacetGrid(df,
               hue = "cut",
               height = 5,
               xlim = (0,20000),
              ).
 map(sns.kdeplot, "price", shade = True).
 add_legend()
);
#FacetGrid=grafik uzerine eklenen boyutlari bolerek gostermek icin
#ustteki kod sunu ifade ediyor;
#price'a gore bir kdeplot grafigi olusturagiz fakat FacetGrid kullanarak tum
#"cut" degiskeninin siniflariyla boyutlandirmis olacagiz.
#bunu yapmanin yolu da map fonksiyonu
#add legend, cut degiskeninin siniflarinin(ideal,premium vs.) yanda gozukmesini
#saglayarak graifigin bilgilendirmesini artiriyor
```

```python
#bir ustteki grafik ile bu grafigin verdigi bilgi arasinda daglar kadar fark var
```

Grafik sunu anlatiyor;
"ideal" sinifi yuksek kaliteli siniflardan birisi oldugu icin pahali olmasi beklenmesine ragmen oyle olmadigini, aksine 1000-3000 dolar arasında cok fazla oldugunu gosteriyor.


```python
sns.catplot(x="cut", y="price", hue="color", kind = "point", data = df);
#x ekseni cut degiskeni, y ekseni price degiskeni olan bir grafige, 
#elmaslarin rengini ifade eden, verisetinde bir categorik degisken olan "color"u bir boyut olarak ekle,
#grafigin kindını(turunu) point(nokta) yap.
```

Fiyatin farkli seviyelerinde hem cut'in hem de color'un farkli etkileri oldugunu rahatlikla gozlemliyoruz. Mesela hem premium kesim olan hem de J rengi olan elmaslar en yuksek fiyatta, ideal kesim olmasina ragmen D ve E rengi olan elmaslar en ucuz.

Premium sinifinda rengin onemi daha da artmis, rengine gore 3500 birimden 7000 birime kadar elmas var, ancak fair sinifnda renk elbette biraz fark yaratsa da o kadar farki olmamis fiyat araligi 3500 le 5000 arasinda sikismis.
