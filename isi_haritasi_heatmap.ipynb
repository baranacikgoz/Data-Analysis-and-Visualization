## Isı Haritası Heatmap

Farklı veri (sadece bu seferlik, heatmap için)


```python
import seaborn as sns
flights = sns.load_dataset("flights")
df = flights.copy()
df.head()
```

```python
df.dtypes
```

```python
df["passengers"].describe()
```

```python
sns.heatmap(df);
```

```python
#Hata aldik cünkü heatmap kullanabilmek icin; verisetini
#daha once ogrendigimiz pivot tablo islemine tabi tutmamiz gerekir.
#daha yapisal hale getirmeliyiz.
```

```python
df #onceki hali
```

```python
df = df.pivot("month", "year", "passengers");
```

```python
df #pivot tablo ile yeni ve cok daha okuna<kli hali
```

```python
sns.heatmap(df);
```

```python
#yillar arttikca genel bir yolcu sayisi artisi var,
#her yili kendi icinde degerlendirince de ozellikle yaz aylarinda
#daha fazla seyahat var.
```

```python
sns.heatmap(df, annot = True, fmt = "d");
```

```python
sns.heatmap(df, annot = True, fmt = "d", linewidths = 0.5);
```
